## Método de bisección Ejercicio I
#### Se buscan las raíces de la ecuación $f(x) = x^3 - 7x^2 + 14x - 6 =0$ en el intervalo [0,1],[1,3.2], [3.2,4] y graficar la función donde se encuentra cada raíz 
    

## **Para el primer intervalo**

```python
import numpy as np
# a y b son los límites del intervalo  
    
#INTERVALO I a = 0 b = 1  [0,1] Proceso exitoso. Si hay raíz

a = 0
b = 1



TOL = 0.01  #TOL es el patámetro de tolerancia para la convergencia del algoritmo

Nmax = 100   #Nmax es el número máximo de iteraciones
```

```python
i = 1     #Contador del bucle
FA = a**3 - 7*a**2+14*a-6    #F(a)
FP = 100                #F(p)
exito = False
p = 10
xp = []  # Aproximaciones a "p" en cada iteración
Funp = []  # Evaluación de xp
ite = []   #Número de la iteración
```

### Se realiza el proceso iterativo 

```python
while i <= Nmax:
    p = a + (b-a)/2.00  # Calculamos el punto medio
    FP = p**3 + 7*p**2 +14*p-6
    xp = np.append(xp,p)
    Funp = np.append(Funp, FP)
    ite = np.append(ite, i)
    print(i,a,b,p,FP)
    if(FP==0):   #¿Es solución?   #abs(FP)< TOL
        print("LA raíz es:")
        print(p)
        exito = True
        break   # Salir del ciclo
    elif(FA*FP > 0):
        a = p
        FA = FP
    else:
        b = p
    i = i+1
    
if exito == False:
    print("Proceso no exitoso")
```

### Se analiza graficamente 

```python
import matplotlib.pyplot as plt
plt.plot(xp)
plt.ylabel('Solución')
plt.xlabel('iteración')
plt.xlim(0, 10)
```
