## Funci√≥n : $\frac{2 - e^{-x} + x^2}{3}$

```python
 from math import *
def g(x):
    x=(2-e**(-x) +x**2)/3
    return x
p0=1
tol=0.00001
N_ite=100
p=10
i=1
while i<=N_ite:
    p=g(p0)
    if abs(p-p0)<=tol:
        print("El punto fijo es ",p," despues de ",i," iteraciones")
        break
    i=i+1
    p0=p
if i>=n0:
    print("El metodo no converge ")
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-5,5,0.1)
y = (2-e**-x +x**2)/3
zeros = np.zeros(len(x))
plt.plot(x,y,'g',x,zeros, '--r')
plt.xlabel('x')
plt.ylabel('y')
```

## Funci√≥n : $0.5(sin(x) - cos(x))$

```python
from math import *
import numpy as np
def g(x):
    x=(0.5 *(math.sin(x) - math.cos(x)))
    return x
p0=1
tol=0.00001
N_ite=50
p=10
i=1
while i<=N_ite:
    p=g(p0)
    if abs(p-p0)<=tol:
        print("El punto fijo es ",p," despues de ",i," iteraciones")
        break
    i=i+1
    p0=p
if i>=n0:
    print("El metodo no converge ")
```

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(-10,10,0.1)
y = 0.5*(np.sin(x) - np.cos(x))
zeros = np.zeros(len(x))
plt.plot(x,y,'g',x,zeros, '-.r')
plt.xlabel('x')
plt.ylabel('y')
```
