```python
from mpl_toolkits.mplot3d import axes3d
import matplotlib.pyplot as plt
import numpy as np

fig = plt.figure()

ax = Axes3D(fig)

x = np.linspace(0, 10, 100)
y = np.linspace(0, 10, 100)
z = np.cos(x) + y * np.sin(x)

#mtodos para rotular
ax.set_xlabel("$x$")
ax.set_ylabel("$y$")
ax.set_zlabel("$z$")
ax.set_title("$z = cos(x) + y \ sen(x)$")

#mtodo para configurar la vista
ax.view_init(20, 75)

ax.plot(x, y, z, color='r')

plt.show()
```

```python
def chicharronera (a,b,c):
    if (b**2-4*a*c) <0 :
        return "No hay solucion"
    else :
        sf = ((b**2-4*a*c))**0.5
        x_1 = (-b +sf)/2*a
        x_2= (-b-sf)/2*a
        return (x_1,x_2)
```
