### Bucles

#### *While*

```python
while True:   #Bucle infinito
    print("Hola")
```

```python
edad = 0
while edad <= 5:
    print("Tienes: " + str(edad))
    edad = edad+1
```

#### *For*

```python
lista = [ "Elemento 1","Elemento 2", "Elemento 3"]
for elem in lista:  #elem sólo es valido en el for y puede tener cualquier nombre
    print(elem)
```

```python
cadena = "Hiram"
for letra in cadena:
    print(letra)
```

Puede usarse para cualquier elemento de tipo secuencia

```python
import numpy as np
```

```python
array1 = np.zeros(9)
array2 = np.array([(1,2),(9,8)])
for numero in array2, array1:
    print(numero)
    
```
