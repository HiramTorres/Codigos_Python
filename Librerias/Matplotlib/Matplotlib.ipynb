```python
import matplotlib.pyplot as plt

plt.figure(figsize = (10,6)) 
x = [2016,2017,2018,2019,2020,2021,2022]
y = [39,41,43,44,47,50,51]
w = [35,31,33,34,37,30,31]
z = [2016,2017,2018,2019,2020,2021,2022]

plt.plot(z,w,marker = "o", color = "m",linestyle = "--",label = "silao")
plt.plot(x,y,marker = "s", color = "c",linestyle = "--",label = "León")
plt.xlabel("Año")
plt.ylabel("Crecimiento por millón")
plt.title("Crecimiento poblacional México")
plt.legend(loc = "center right")
# plt.savefig("grafico.png")
#plt.yticks([30,40,50]) Para solo colocar ciertos numeros en alguno de los ejes ya se a x o y

plt.show()
```

## Subplots 


```python
fig, ax = plt.subplots(1,2, sharey = True)
ax[0].plot(x,y,color = 'c',label = 'León')
ax[0].legend()
ax[1].plot(z,w,color = 'm',label = 'silao')
ax[1].legend()
plt.show()
```

## Bar plots

```python
x1 = ["argentina", "Colombia", "Perú"]
y1 = [50,39,46]

plt.plot(x1,y1,color = "r", linestyle = "--", marker = "o")
plt.bar(x1,y1,color = "c")
plt.show()
```

## Piecharts

```python
plt.pie(y1, labels = x1) # Primero el dato numerico
plt.show()
```

## Histogram



```python
import random
edades = [random.randint(0,51) for x in range(50)]
bins = [15,25,35,40,50]

plt.hist(edades,bins, edgecolor = "black")
plt.show()
```

## Boxplots


```python
plt.boxplot(edades)
plt.show()
```

## scatterplots

```python
a = [random.randrange(0,100) for x in range(100)]
b = [random.randrange(0,100) for y in range(100)]

plt.scatter(b,a, color = "b")
plt.scatter(a,b, color = "g")
plt.show()
```
